```python
# first trial (효율성 실패)
from collections import defaultdict
def solution(info, query):
    infinite_dict=lambda: defaultdict(infinite_dict)
    record, ans = infinite_dict(), []
    for i in info:
        tmp = i.split()
        if record[tmp[0]][tmp[1]][tmp[2]][tmp[3]]['score']:
            record[tmp[0]][tmp[1]][tmp[2]][tmp[3]]['score'].append(int(tmp[4]))
        else:
            record[tmp[0]][tmp[1]][tmp[2]][tmp[3]]['score'] = [int(tmp[4])]
            
    for i in query:
        tmp, cnt, depth = i.split()[::2] + [int(i.split()[-1])], [], 0
        dfs(record,tmp[0], tmp, cnt, depth)
        ans.append(sum(cnt))
    return ans

def dfs(record, cur, query, cnt, depth):
    if depth == 4:
        cnt.append(sum(map(lambda x: x >= cur, record['score'])))
        return
    elif cur == '-':
        for i in record:
            dfs(record[i], query[1], query[1:], cnt, depth+1)
    else:
        dfs(record[cur], query[1], query[1:], cnt, depth+1)
        
        
info = ["java backend junior pizza 150",
    "python frontend senior chicken 210",
    "python frontend senior chicken 150",
    "cpp backend senior pizza 260",
    "java backend junior chicken 80",
    "python backend senior chicken 50"]

query = ["java and backend and junior and pizza 100",
         "python and frontend and senior and chicken 200",
         "cpp and - and senior and pizza 250",
         "- and backend and senior and - 150",
         "- and - and - and chicken 100",
         "- and - and - and - 150"]
solution(info, query)
```

```python
# second trial (using 이분탐색)
from collections import defaultdict
from itertools import combinations
import re
def solution(info, query):
    record, ans = defaultdict(list), []
    for i in info:
        tmp, score = i.split()[:-1], int(i.split()[-1])
        for j in range(0, 5):
            [record['all'].append(score) if not k else record[''.join(k)].append(score) for k in combinations(tmp, j)]
    [values.sort() for key, values in record.items()]

    for i in query:
        tmp, score = i.split()[::2], int(i.split()[-1])
        requirements=re.sub('[-]','',''.join(tmp))
        check = record['all'] if not requirements else record[requirements]
        left, right = 0, len(check)-1
        while left <= right:
            mid = (left+right)//2
            if check[mid] >= score:
                right = mid - 1
            else:
                left = mid + 1
        ans.append(len(check)-left)
       
    return ans

info = ["java backend junior pizza 150",
    "python frontend senior chicken 210",
    "python frontend senior chicken 150",
    "cpp backend senior pizza 260",
    "java backend junior chicken 80",
    "python backend senior chicken 50"]

#query = ["- and - and - and - 150"]

query = ["java and backend and junior and pizza 100",
         "python and frontend and senior and chicken 200",
         "cpp and - and senior and pizza 250",
         "- and backend and senior and - 150",
         "- and - and - and chicken 100",
         "- and - and - and - 150"]
solution(info, query)
```

```python
# 3rd trial (using insort, bisect_left)
from collections import defaultdict
from itertools import combinations
from bisect import insort, bisect_left
import re
def solution(info, query):
    record, ans = defaultdict(list), []
    for i in info:
        tmp, score = i.split()[:-1], int(i.split()[-1])
        for j in range(0, 5):
            [insort(record['all'],score) if not k else insort(record[''.join(k)],score) for k in combinations(tmp, j)]

    for i in query:
        tmp, score = i.split()[::2], int(i.split()[-1])
        requirements=re.sub('[-]','',''.join(tmp))
        check = record['all'] if not requirements else record[requirements]
        ans.append(len(check)-bisect_left(check, score))
    return ans
    
info = ["java backend junior pizza 150",
    "python frontend senior chicken 210",
    "python frontend senior chicken 150",
    "cpp backend senior pizza 260",
    "java backend junior chicken 80",
    "python backend senior chicken 50"]

#query = ["- and - and - and - 150"]

query = ["java and backend and junior and pizza 100",
         "python and frontend and senior and chicken 200",
         "cpp and - and senior and pizza 250",
         "- and backend and senior and - 150",
         "- and - and - and chicken 100",
         "- and - and - and - 150"]
solution(info, query)
```

```python
# 5th trial (다른 아이디어)
from collections import defaultdict
from bisect import bisect_left, insort
def solution(info, query):
    answer = []
    graph = defaultdict(list)
    
    for person in map(lambda x:x.split(), info):
        insort(graph[tuple(person[:-1])], int(person[-1]))
        
    for require in map(lambda x:x.replace('and', '').split(), query):
        stack = list(graph.keys())
        for idx, what in enumerate(require[:-1]):
            if what != '-':
                stack = [temp for temp in stack if temp[idx] == what]
        answer.append(sum([len(graph[x]) - bisect_left(graph[x], int(require[-1])) for x in stack]))
    return answer



info = ["java backend junior pizza 150",
    "python frontend senior chicken 210",
    "python frontend senior chicken 150",
    "cpp backend senior pizza 260",
    "java backend junior chicken 80",
    "python backend senior chicken 50"]

#query = ["- and - and - and - 150"]

query = ["java and backend and junior and pizza 100",
         "python and frontend and senior and chicken 200",
         "cpp and - and senior and pizza 250",
         "- and backend and senior and - 150",
         "- and - and - and chicken 100",
         "- and - and - and - 150"]
solution(info, query)
```
